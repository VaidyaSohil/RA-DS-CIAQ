```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = [16,9]
```

```python
# (pm2.5 >10) = 1
# (pm2.5 <= 10) = 0

# Class Distribution:
# 0    6250
# 1    1023

df = pd.DataFrame({'parameters':['accuracy', 'precision 0', 'precision 1', 'recall 0', 'recall 1'], 'score':[0.86, 0.86, 0.50, 0.99, 0.05]})
ax = df.plot.bar(x='parameters', y='score', rot=0, fontsize = 14)
```

```python
# Key takeaway: (2 classes)
# - Uneven distribution (pm2.5 > 10) = Class 1 [1023]; (pm2.5 <= 10) = Class 0 [6250]
# - Recall (Higher False Negative for class 1)
```

```python
# I.
# (pm2.5 > 4) = 1
# (pm2.5 <= 4) = 0

# Class Distribution:
# 0    3790
# 1    3483

accuracy = [0.55, 0.60, 0.69, 0.55, 0.64, 0.69, 0.59]
precision0 = [0.57, 0.62, 0.68, 0.56, 0.67, 0.70, 0.62]
precision1 = [0.53, 0.58, 0.69, 0.53, 0.62, 0.68, 0.57]
recall0 = [0.62, 0.64, 0.76, 0.67, 0.64, 0.72, 0.59]
recall1 = [0.48, 0.57, 0.60, 0.42, 0.64, 0.66, 0.59]

index = ['TECH A - TF_CNT |TEMP', 'TECH A - TF_CNT |TEMP |SPD |DIR |PRES', 'TECH A - TF_CNT |TEMP |SPD |DIR |PRES |NO2 |Month', 'TECH B - TF_CNT |TEMP', 'TECH B - TF_CNT |TEMP |SPD |NO2 |DIR', 'TECH C - TF_CNT |TEMP |Month |Hour', 'TECH C - TF_CNT |TEMP |NO2 |SPD |DIR |Month |Hour']
df = pd.DataFrame({'accuracy': accuracy, 'precision 0': precision0, 'precision 1': precision1, 'recall 0':recall0, 'recall 1':recall1}, index=index)
ax = df.plot.bar(rot=60, fontsize = 14)
ax.set_title("Class 0 (PM2.5 <= 4) vs Class 1 (PM2.5 > 4)", fontsize=25)
ax.set_xlabel("Paramters", fontsize=20)
ax.set_ylabel("Score", fontsize=20)
```

```python
# Key Take Away (2 classes)
# - 3 Techniques: A. Gaussian B. Categorical C. Gaussian + Categorical
# - Better class distribution than previous attempt (pm2.5 > 4) = Class 1[3483] (pm2.5 <= 4) = Class 0[3790]
# - Technique A performed the joint-best (Including sufficient variables)
# - Technique B performed well with inclusion of sufficient variables
# - Technique C performed the joint-best (Including just TF_CNT and TEMP as continuous; Month and Hour as discrete)
# - Technique C performed worse with inclusion of other variables like NO2, SPD, DIR

# Previous attempts were biased and resulted in about 85% accuracy, whereas these equal distributions perform at
# around 60-70 % accuracy range. Including and Excluding different variable changes the performance of the model
```
